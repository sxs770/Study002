# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_csv('score.csv', index_col = '지원번호')
df
```

## column 수정

```python
df['학교'].replace({'북산고' : '상북고', '능남고' : '남능고'})
```

```python
df #inplace를 해줘야한다 if 값을 바꾸고 싶을 때
```

```python
df['학교'].replace({'북산고' : '상북고'}, inplace = True)
```

```python
df['SW특기'] = df['SW특기'].str.lower()
```

```python
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
```

 - str.lower() : 문자열을 모두 소문자로 변경
 - str.upper() : 문자열을 모두 대문자로 변경
 - str.capitalize() : 문자열의 앞 글자는 대문자로, 나머지 글자는 소문자로 변경

```python
df
```

```python
df['학교'] = df['학교'] + '등학교'
```

```python
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
```

```python
df
```

```python
df['결과'] = 'Fail' # 결과 Column을 추가하고 전체 데이터는 Fail로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' #총합이 400보다 큰 데이터에 대해서 결과를 Pass로 업데이트
```

```python
df
```

## Column 삭제

```python
df.drop(columns = ['총합']) #총합 Column 삭제
```

```python
df.drop(columns = ['국어', '영어', '수학']) #국어, 영어, 수학 삭제
```

```python
df #dataframe에 반영하려면 inplace 필요
```

## Row 삭제

```python
df.drop(index = '4번') # 4번 학생 데이터 Row 삭제
```

```python
filt = df['수학'] < 80 #수학 점수가 80점 미만 학생 필터링
df[filt]
```

```python
df[filt].index
```

```python
df.drop(index = df[filt].index) #수학 점수가 80점 미만 학생 삭제
```
