## 데이터 제거하기 dropna

```python
df = pd.read_csv('score.csv', index_col = '지원번호')
df
```

```python
df.dropna(inplace = True) #NaN을 포함하는 데이터 삭제
```

```python
df
```

```python
df = pd.read_csv('score.csv', index_col = '지원번호')
df
```

axis : index or column
 - low를 지울지 columns를 지울지를 선택하는 옵션  
 
how : any or all  
 - any : 데이터 중에 하나라도 결측치가 있을 경우 삭제
 - all : 데이터가 모두 결측치일 경우 삭제

```python
df.dropna(axis = 'index', how = 'any') #NaN이 하나라도 있는 row 삭제
```

```python
df.dropna(axis = 'columns') #NaN이 하나라도 있는 column 삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis = 'columns', how = 'all') #데이터 전체가 NaN인 경우에만 Column 삭제
```
